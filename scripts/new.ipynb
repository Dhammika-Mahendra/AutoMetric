```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('../data/car_price.csv')

# Print all column names
print(df.columns)
```

```python
#print all unique values in the second column
print(df.iloc[:, 1].unique())
```

```python
#print all unique values in the sixth column into a new file by sorting ascending order
unique_values = sorted(df.iloc[:, 8].unique())
with open('unique_values.txt', 'w') as f:
    for value in unique_values:
        f.write(f"{value}\n")
```
